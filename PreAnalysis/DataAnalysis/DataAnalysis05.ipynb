## Inference

When charts are examined, it can be noticed that some categories such as Shirt or Accessory with the ProductType feature are very few in the dataset.
This is an undesirable case.

This case is one of the most important reasons why a lot of data is needed in machine learning methods.
It is obviously unrealistic to expect any dataset to be balanced for each feature.
However, the number of representatives of each value should be at least enough to generalized sufficiently by the model.

For example, there is only one Accessory value of the ProductType feature in the entire dataset.
In this case, if this single row is used for training, it cannot be sufficiently generalized by the model and cannot be tested in any way because it is not representative in the test data.
If used for testing, the model will try to predict on a value that it has never encountered even a representative of it before.
