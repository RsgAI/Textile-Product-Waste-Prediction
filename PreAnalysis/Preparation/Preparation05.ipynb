```python
#Now the values in the ProductDetail column in the dataFrame will be processed for extracting more features from it

#productDetailFrame that contains the detailed information of the values in the dataFrame ProductDetail column is being created
productDetailFrame = pd.DataFrame({
                            'ProductDetail' : dataFrame['ProductDetail'].unique()
                            })
productDetailFrame
```

```python
#The function that will extract the Maturity information from the ProductDetail column is being defined
def ExtractMaturity(productDetail):
    #Çocuk means Child in Turkish
    if 'COCUK' in productDetail:
        return 'Child'
    #Bebek means Baby in Turkish
    elif 'BEBEK' in productDetail:
        return 'Baby'
    #Others are Adult
    else:
        return 'Adult'
```

```python
#ExtractMaturity function is being applied for each value in the ProductDetail column
#and the results are being written to the Maturity column
productDetailFrame['Maturity'] = productDetailFrame['ProductDetail'].apply(lambda x : ExtractMaturity(x))
productDetailFrame
```

```python
#The function that will extract the Gender information from the ProductDetail column is being defined
def ExtractGender(productDetail):
    #Kadın and Kız denote female gender in Turkish
    if 'KADIN' in productDetail or 'KIZ' in productDetail:
        return 'Female'
    #Erkek denote male gender in Turkish
    elif 'ERKEK' in productDetail:
        return 'Male'
    #Others will be considered as Unisex
    else:
        return 'Unisex'
```

```python
#ExtractGender function is being applied for each value in the ProductDetail column
#and the results are being written to the Gender column
productDetailFrame['Gender'] = productDetailFrame['ProductDetail'].apply(lambda x : ExtractGender(x))
productDetailFrame
```

```python
#The function that will extract the ProductType information from the ProductDetail column is being defined
def ExtractProductType(productDetail):
    if 'T-SHIRT' in productDetail:
        return 'T-Shirt'
    #Pantolon means Pant in Turkish
    elif 'PANTOLON' in productDetail:
        return 'Pant'
    #Gömlek means Shirt in Turkish
    elif 'GOMLEK' in productDetail:
        return 'Shirt'
    #Elbise means Dress in Turkish
    elif 'ELBISE' in productDetail:
        return 'Dress'
    #Pijama  means Pyjamas in Turkish
    elif 'PIJAMA' in productDetail:
        return 'Pyjamas'
    #Etek means Skirt in Turkish
    elif 'ETEK' in productDetail:
        return 'Skirt'
    elif 'SWEAT' in productDetail:
        return 'Sweatshirt'
    #Aksesuar means Accessory in Turkish
    elif 'AKSESUAR' in productDetail:
        return 'Accessory'
    else:
        return None
```

```python
#ExtractProductType function is being applied for each value in the ProductDetail column
#and the results are being written to the ProductType column
productDetailFrame['ProductType'] = productDetailFrame['ProductDetail'].apply(lambda x : ExtractProductType(x))
productDetailFrame
```

```python
#unique value counts for columns in productDetailFrame are being printed
productDetailFrame.nunique()
```

```python
#When the unique value counts are examined, it can be seen that there are 8 unique values for the ProductType feature
#When the data is examined, it can be noticed that more general categories can be created for these categories

#The function that will extract the ProductTypeCategory information from the ProductType column is being defined
def ExtractProductTypeCategory(ProductType):
    if 'T-Shirt' in ProductType or 'Shirt' in ProductType or 'Sweatshirt' in ProductType:
        return 'Top'
    elif 'Pant' in ProductType or 'Skirt' in ProductType:
        return 'Leg'
    elif 'Dress' in ProductType or 'Pyjamas' in ProductType:
        return 'Full'
    else:
        return 'Other'
```

```python
#ExtractProductTypeCategory function is being applied for each value in the ProductType column
#and the results are being written to the ProductTypeCategory column
productDetailFrame['ProductTypeCategory'] = productDetailFrame['ProductType'].apply(lambda x : ExtractProductTypeCategory(x))
productDetailFrame
```

```python
#Since all information about ProductDetails is kept in the productDetailFrame,
#it will bi sufficient to keep foreign key values of ProductDetails representing the productDetailFrame data in the dataFrame

#dataFrame ProductDetail column is being recreated as productDetailFrame indexes
dataFrame['ProductDetail'] = dataFrame['ProductDetail'].apply(
    lambda x : productDetailFrame.index[productDetailFrame['ProductDetail'] == x].tolist()[0])
dataFrame.head()
```

```python
#ProductDetail column is being renamed as ProductDetailID
dataFrame.rename(columns = {'ProductDetail' : 'ProductDetailID'}, inplace = True)
dataFrame.head()
```
