FabricCode and FabricDetail features will be examined in depth.

```python
#A pandas dataframe named as dataFrame is being created by reading the data pkl file created last process
dataFrame = pd.read_pickle("../Data/WithRelationalTables/FabricWasteProductDetail.pkl")
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataframe is being printed
dataFrame.describe()
```

```python
#Number of unique values of each column are being printed
dataFrame.nunique()
```

```python
#The FabricCode feature can be thought of as the ID of the FabricDetail feature
#In this case, the number of unique values must be equal
#However, when the unique valu numbers examined, it can be seen that
#the FabricCode column has 9 unique values and
#the FabricDetail column has 12 unique values
#This is an unexpected situation and needs to be investigated

#FabricDetail values are being printed for each FabricCode value
pd.DataFrame(dataFrame.groupby(['FabricCode', 'FabricDetail']).size(), columns=['NumberOfRow'])
```

```python
#When the data is examined, it can be seen that there are 4 different FabricDetail values for the 2RB201 FabricCode value
#Although they look the same when printed in DataFrame,
#there must be differences hidden by formatting method used in printing process

#FabricDetail values, which is FabricCode 2RB201, are being printed in its raw form for investigate differences
dataFrame[dataFrame['FabricCode'] == '2RB201']['FabricDetail'].values
```

```python
#When the results are examined, it can be seen that the difference is due to the whitespaces between characters
#To solve the problem, multiple whitespaces next to each other are made into singele whitespaces 
#then leading and trailing whitespaces are deleted
#Since this process will transform the data smoother form without breaking its structure,
#it is being applied not only the rows with FabricCode 2RB201, bu to entire column
dataFrame['FabricDetail'] = dataFrame['FabricDetail'].replace(r'\s+', ' ', regex = True).str.strip()
dataFrame.head()
```

```python
#FabricDetail values, which is FabricCode 2RB201, are being printed in its raw form
dataFrame[dataFrame['FabricCode'] == '2RB201']['FabricDetail'].values
```

```python
#FabricDetail values are being printed for each FabricCode value
pd.DataFrame(dataFrame.groupby(['FabricCode', 'FabricDetail']).size(), columns=['NumberOfRow'])
```

```python
#fabricDetailFrame containing FabricCode - FabricDetail pairs is being created 
fabricDetailFrame = pd.DataFrame(dataFrame.groupby(['FabricCode', 'FabricDetail']).size()).reset_index().drop([0], axis = 1)
fabricDetailFrame
```

```python
#The function that will extract the FabricType information from the FabricDetail column is being defined
def ExtractFabricType(fabricDetail):
    if 'SUPREM' in fabricDetail:
        return 'Single Jersey'
    elif 'RIBANA' in fabricDetail:
        return 'Ribana'
    elif 'INTERLOK' in fabricDetail:
        return 'Interlock'
```

```python
#ExtractFabricType function is being applied for each value in the FabricDetail column
#and the results are being written to the FabricType column
fabricDetailFrame['FabricType'] = fabricDetailFrame['FabricDetail'].apply(lambda x : ExtractFabricType(x))
fabricDetailFrame
```

```python
#Actually more features can be extracted from FabricDetail column
#Since this notebook file is part of the preliminary analysis, a more detailed analysis will not be done in this notebook

#FabricDetail column is being removed from dataFrame
dataFrame.drop(['FabricDetail'], axis = 1, inplace = True)
dataFrame.head()
```
