# First Data AnalysisProcess

In this notebook file FabricWaste DataFrame will be merged with related DataFrames and saved as pkl file for Data Analysis prcesses

```python
#Library is being imported
import pandas as pd
```

```python
#Library version is being printed
print('pandas Version: ' + pd.__version__)
```

```python
#A pandas dataframe named as dataFrame is being created by reading the data pkl file created last process
dataFrame = pd.read_pickle("../Data/FinalPreparation/FabricWaste.pkl")
dataFrame.head()
```

```python
#colorFrame is being created from pkl file
colorFrame = pd.read_pickle("../Data/WithRelationalTables/Color.pkl")
colorFrame.head()
```

```python
#For each row in the dataFrame, the data in the colorFrame is being added to dataFrame
#This operation is being done by merging dataFrame and colorFrame on ColorID
dataFrame = pd.merge(dataFrame, colorFrame, left_on = 'ColorID', right_index = True, how = 'left')
dataFrame.head()
```

```python
#Color, ColorTypeName and ColorCode columns that saved for informational purposes are being removed from dataFrame
dataFrame.drop(['Color', 'ColorTypeName', 'ColorCode'], axis = 1, inplace = True)
dataFrame.head()
```

```python
#productDetailFrame is being created from pkl file
productDetailFrame = pd.read_pickle("../Data/WithRelationalTables/ProductDetail.pkl")
productDetailFrame.head()
```

```python
#For each row in the dataFrame, the data in the productDetailFrame is being added to dataFrame
#This operation is being done by merging dataFrame and productDetailFrame on ProductDetailID
dataFrame = pd.merge(dataFrame, productDetailFrame, left_on = 'ProductDetailID', right_index = True, how = 'left')
dataFrame.head()
```

```python
#ProductDetail column that saved for informational purposes is being removed from dataFrame
dataFrame.drop('ProductDetail', axis = 1, inplace = True)
dataFrame.head()
```

```python
#fabricDetailFrame is being created from pkl file
fabricDetailFrame = pd.read_pickle("../Data/WithRelationalTables/FabricDetail.pkl")
fabricDetailFrame.head()
```

```python
#For each row in the dataFrame, the data in the fabricDetailFrame is being added to dataFrame
#This operation is being done by merging dataFrame and fabricDetailFrame on FabricCode
dataFrame = dataFrame.merge(fabricDetailFrame, on = 'FabricCode', how = 'left')
dataFrame.head()
```

```python
#FabricDetail column that saved for informational purposes is being removed from dataFrame
dataFrame.drop('FabricDetail', axis = 1, inplace = True)
dataFrame.head()
```

```python
#ColorID, ProductDetailID, FabricCode columns 
#that only used to relate colorFrame, productDetailFrame amd fabricDetailFrame to dataFrame are being removed
dataFrame.drop(['ColorID', 'ProductDetailID', 'FabricCode'], axis = 1, inplace = True)
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#dataFrame columns are being reordered in CategoricalInput/ContinuousInput/Output format
columnOrder = ['ProductTypeCategory', 'ProductType', 'Maturity', 'Gender', 'FabricType', 'ColorType', 'CustomerDefinedCategory',
               'IsManualProcess', 'Red', 'Green', 'Blue', 'Pus', 'Fine', 'G/M2', 'Cotton', 'Nylon', 'Fiber', 'Linen', 'Hemp',
               'Weight', 'Width', 'ProductPart', 'NotchLayer', 'NotchQuantity', 'OrderQuantity',
               'PrintErrorRate', 'SewingErrorRate', 'FabricStainRate', 'FabricErrorRate', 'EmbroideryErrorRate',
               'MeasureErrorRate', 'SecondQualityRate', 'DefectRate', 'PrintError', 'SewingError', 'FabricStain', 'FabricError',
               'EmbroideryError', 'MeasureError', 'SecondQuality', 'CalculatedSecondQuality', 'Defect']
dataFrame = dataFrame[columnOrder]
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataframe is being printed
dataFrame.describe()
```

```python
#dataFrame is being saved as pkl files
dataFrame.to_pickle("../Data/DataAnalysis/FabricWaste.pkl")
```
