In this notebook file, countplots will be plotted for CustomerDefinedCategory categorical feature, and boxplots will be plotted to check the output rate distributions of the categories

```python
#Libraries are being imported
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import __version__ as matplotlibVersion
import seaborn as sns
```

```python
#Settings for Graphic Libraries are being made
%matplotlib inline
sns.set_style('whitegrid')
```

```python
#Library versions are being printed
print('pandas Version: ' + pd.__version__)
print('matplotlib Version: ' + matplotlibVersion)
print('seaborn Version: ' + sns.__version__)
```

```python
#A pandas dataframe named as dataFrame is being created by reading the data pkl file created last process
dataFrame = pd.read_pickle("../Data/DataAnalysis/FabricWaste.pkl")
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataframe is being printed
dataFrame.describe()
```

```python
#Chart that showing distribution of CustomerDefinedCategory categorical feature values in dataFrame is being plotted
plt.figure(figsize = (12, 8))
columnOrder = dataFrame['CustomerDefinedCategory'].value_counts().index
sns.countplot(x = 'CustomerDefinedCategory', data = dataFrame, order = columnOrder)
```

```python
#A boxplot, to check the PrintErrorRate distributions for each CustomerDefinedCategory value, is being plotted
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'CustomerDefinedCategory', y = 'PrintErrorRate', data = dataFrame, order = columnOrder)
```
