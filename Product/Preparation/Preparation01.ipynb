# First Data Preparation Process

In this notebook file, the dataset provided by the customer will be eddited for general purposes.

Features that are meaningless or unusable will be removed from the dataset.

```python
#Library is being imported
import pandas as pd
```

```python
#Library version is being printed
print('pandas Version: ' + pd.__version__)
```

```python
#A pandas dataframe named as dataFrame is being created by reading the data from an excel file
dataFrame = pd.read_excel('../Data/RawData/FabricWaste.xlsx')
dataFrame.head()
```

```python
#Information of dataFrame is being printed
dataFrame.info()
```

```python
#Statistical information of dataFrame is being printed
dataFrame.describe()
```

```python
#Since OrderNo is an id value for the relevant order, it is not meaningful to use this value as input or output
#OrderNo is being removed from the dataset as it does not make any sense in this project
dataFrame.drop('OrderNo', axis = 1, inplace = True)
dataFrame.head()
```

```python
#The Planned Column is the number of planned productions to be able to deliver number of product OrderQuantity
#It will not be used, as the expectation within the scope of this project is to make more meaningful planning
#For this reason, the relevant column is being removed
dataFrame.drop('Planned', axis = 1, inplace = True)
dataFrame.head()
```

```python
#Looking at the Quantity and Print columns, it is seen that they frequently take the same value
#For this reason, it is being checked whether they always get the same value
dataFrame[dataFrame['Quantity'] != dataFrame['Print']].shape
```

```python
#As can be see, there is no row satisfying the Quantity != Print Condition
#According to the dataset, these two columns mean the same feature
#One of the relevant columns is being removed
dataFrame.drop('Print', axis = 1, inplace = True)
dataFrame.head()
```

```python
#When information of dataFrame is examined, it  seen that there are only 61 non-null values for the Reason 1 column
#Since the column take string values and contain a large number of null values,
#it seems very difficult to use it by handling it

#However, given that this attribute is a description attribute
#that explains the numerical difference between Planned and Quantity,
#It can be seen that any action to be taken on this attribute is already meaningless.

#Reason1 column is being removed
dataFrame.drop('Reason1', axis = 1, inplace = True)
dataFrame.head()
```

```python
#Similarly, the Reason2 attribute is a description attribute
#that explains the dominant reason why second quality products are second quality
#This attribute is unnecessary since the reasons for the second quality products are detailed in the dataFrame

#Reason2 column is being removed
dataFrame.drop('Reason2', axis = 1, inplace = True)
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataFrame is being printed
dataFrame.describe()
```

```python
#dataFrame cleaned of columns that will not be used is being saved as pkl file
dataFrame.to_pickle("../Data/NeddedColumns/FabricWaste.pkl")
```
