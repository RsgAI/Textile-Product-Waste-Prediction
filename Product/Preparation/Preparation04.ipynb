```python
#A pandas dataframe named as colorFrame is being created by reading the data from an excel file provided by customer
colorFrame = pd.read_excel('../Data/RawData/ColorCategory.xlsx')
colorFrame
```

```python
#dataFrame and colorFrame are being merged on Color column
mergedFrame = pd.merge(dataFrame, colorFrame, how='left', on=['Color'])
mergedFrame.head()
```

```python
#In this section, it should be examined 
#whether the colors that are not in the Color Frame resemble one of the colors in the Color Frame

#Unique color values in colorFrame are being printed
colorFrame['Color'].unique()
```

```python
#Processes are being performed according to the color values of the rows in the mergedFrame 
#whose CustomerDefinedCategory value is null and the unique color values in the colorFrame

#AUBERGINE 2 color values in dataFrame are being converted to AUBERGINE color value in colorFrame
dataFrame.loc[dataFrame['Color'] == 'AUBERGINE 2', 'Color'] =  'AUBERGINE'

#BLASGELB 1 color values in dataFrame are being converted to BLASGELB color value in colorFrame
dataFrame.loc[dataFrame['Color'] == 'BLASGELB   1', 'Color'] =  'BLASGELB'

#GRAU(SNOW) MELANGE V2 color values in dataFrame are being converted to GRAU(SNOW) MELANGE color value in colorFrame
dataFrame.loc[dataFrame['Color'] == 'GRAU(SNOW) MELANGE V2', 'Color'] =  'GRAU(SNOW) MELANGE'

#BLACK/WHITE FINE STRIPED color values in dataFrame are being converted to BLACK/WHITE color value in colorFrame
dataFrame.loc[dataFrame['Color'] == 'BLACK/WHITE FINE STRIPED', 'Color'] =  'BLACK/WHITE'

dataFrame.head()
```

```python
#dataFrame and colorFrame are being merged on Color column
mergedFrame = pd.merge(dataFrame, colorFrame, how='left', on=['Color'])
mergedFrame.head()
```

```python
#Rows without CustomerDefinedCategory information in colorFrame are being printed
mergedFrame[mergedFrame['CustomerDefinedCategory'].isnull()]
```

```python
#Now that the Color columns in the dataFrame and colorFrame are properly matched, new attributes can be added to the colorFrame

#Unique color values in colorFrame are being printed
colorFrame['Color'].unique()
```

```python
#featureFrame that contains the detailed information of the values in the dataFrame Color column is being created

#ColorCode and ColorType is being created manually by examining the color names
#ColorCode column is in RGB format and ColorCodes of rows containing multi color are assigned as [-1, -1, -1]
#With this method, it is being stated that the ColorCode is meaningless for the products that contain multi color
#In this case it can be seen that the ColorCode attribute must be used with ColorType attribute but ColorType can be used alone
featureFrame = pd.DataFrame({
                            'Color' : colorFrame['Color'].unique(),
                            'ColorCode' : [[-1, -1, -1], [-1, -1, -1], [72, 50, 72], [255, 255, 167],
                                           [106, 90, 205], [250, 250, 250], [173, 216, 230], [218, 205, 129], [94, 251, 110],
                                           [255, 127, 80], [200, 162, 200], [-1, -1, -1], [150, 123, 182], [224, 176, 255], 
                                                                                                               [62,180,137],
                                           [238, 235, 227], [-1, -1, -1], [25, 25, 112], [42, 47, 35], [255, 229, 180],
                                           [0, 95, 106], [255, 178, 208], [-1, -1, -1], [-1, -1, -1], [255, 0 ,127], 
                                           [244, 196, 48], [0, 0, 0], [193, 183, 176], [138, 154, 91], [-1, -1, -1],
                                           [-1, -1, -1], [-1, -1, -1], [48, 213, 200], [105, 97, 86],
                                           [105, 43, 43], [44, 64, 83], [54, 81, 148], [5, 2, 3],
                                           [204, 202, 201], [-1, -1, -1], [229, 211, 191], [210, 237, 218],
                                           [-1, -1, -1], [249, 245, 236], [255, 255, 255], [59, 60,54],
                                           [80, 87, 76], [199, 31, 45]
                                          ],
                            'ColorType' :['AOP', 'AOP', 'SC', 'SC', 'SC', 'M', 'SC', 'SC', 'SC', 'SC', 'SC', 'AOP', 
                                          'SC', 'SC', 'SC', 'SC', 'AOP', 'SC', 'SC', 'SC', 'SC', 'SC', 'AOP', 'AOP',
                                          'M', 'SC', 'SC', 'SC', 'SC', 'AOP', 'AOP', 'S', 'SC', 'SC', 'SC', 'SC',
                                          'SC', 'SC', 'M', 'S', 'SC', 'SC', 'S', 'SC', 'SC', 'SC', 'SC', 'SC']
                            })
featureFrame
```

```python
#typeFrame that contains the full name information of the categories in the featureFrame colorType column is being created
typeFrame = pd.DataFrame({
                                'ColorType' : ['AOP', 'SC', 'M', 'S'],
                                'ColorTypeName' :['All Over Print', 'Solid Color', 'Melange', 'Stripe']
                                })
typeFrame
```

```python
#featureFrame and typeFrame are being merged on ColorType column
featureFrame = pd.merge(featureFrame, typeFrame, how='left', on=['ColorType'])
featureFrame
```

```python
#featureFrame and colorFrame are being merged on Color column
colorFrame =  pd.merge(featureFrame, colorFrame, how='left', on=['Color'])
colorFrame
```

```python
#ColorCodes are being splitted into Red, Green, Blue values
colorFrame['Red'] =  colorFrame['ColorCode'].apply(lambda x : x[0])
colorFrame['Green'] =  colorFrame['ColorCode'].apply(lambda x : x[1])
colorFrame['Blue'] =  colorFrame['ColorCode'].apply(lambda x : x[2])
colorFrame
```

```python
#Since all information about Colors is kept in the colorFrame,
#it will bi sufficient to keep foreign key values of Colors representing the colorFrame data in the dataFrame

#dataFrame Color column is being recreated as colorFrame indexes
dataFrame['Color'] = dataFrame['Color'].apply(lambda x : colorFrame.index[colorFrame['Color'] == x].tolist()[0])
dataFrame.head()
```

```python
#Color column is being renamed as ColorID
dataFrame.rename(columns = {'Color' : 'ColorID'}, inplace = True)
dataFrame.head()
```
