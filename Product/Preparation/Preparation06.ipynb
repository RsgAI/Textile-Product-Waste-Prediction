```python
#The FabricCode feature can be thought of as the ID of the FabricDetail feature
#In this case, the number of unique values must be equal
#However, when the unique valu numbers examined, it can be seen that
#the FabricCode column has 9 unique values and
#the FabricDetail column has 12 unique values
#This is an unexpected situation and needs to be investigated

#FabricDetail values are being printed for each FabricCode value
pd.DataFrame(dataFrame.groupby(['FabricCode', 'FabricDetail']).size(), columns=['NumberOfRow'])
```

```python
#column names for fabric Detail Frame are being defined
fabricDetailColumn = ['FabricCode', 'FabricType', 'Pus', 'Fine', 'G/M2', 'Cotton', 'Nylon', 'Fiber', 'Linen', 'Hemp']
fabricDetailColumn
```

```python
#Since the FabricDetail column has no fixed format and it has few unique values
#Writing a function to extract useful information from this column will require more effort than doing it manually
#This is why the process is done manually

#The information extracted from the FabricDetail column for each FabricCode is being written to the Data list
#Fabric with code 2PK070 was used in only one product, it has no Pus - Fine fields
#and is the only fabric with a FabricType value of Pique
#Therefore, it would be more logical to remove a row from the Dataset instead of estimating for this fabric

#In this case, there is no need to add the relevant fabric to the fabricDetailFrame
data = [
            ['2FT052', 'Diagonal', 32, 20, 320, 1., 0., 0., 0., 0.],
            ['2FT098', 'Diagonal', 32, 20, 320, 1., 0., 0., 0., 0.],
            ['2FT304', 'Diagonal', 32, 20, 320, 1., 0., 0., 0., 0.],
            ['2IN001', 'Interlock', 34, 24, 250, 1., 0., 0., 0., 0.],
            ['2IN007', 'Interlock', 34, 24, 250, 0.95, 0.05, 0., 0., 0.],
            #['2PK070', 'Pique', None, None, 230, 1., 0., 0., 0., 0.],
            ['2RB020', 'Ribana', 34, 18, 210, 1., 0., 0., 0., 0.],
            ['2RB201', 'Ribana', 34, 20, 175, 1., 0., 0., 0., 0.],
            ['2SP021', 'Single Jersey', 32, 28, 125, 1., 0., 0., 0., 0.],
            ['2SP029', 'Single Jersey', 32, 28, 155, 1., 0., 0., 0., 0.],
            ['2SP033', 'Single Jersey', 32, 22, 190, 1., 0., 0., 0., 0.],
            ['2SP050', 'Single Jersey', 32, 28, 125, 0.96, 0., 0.04, 0., 0.],
            ['2SP061', 'Single Jersey', 32, 28, 190, 0.96, 0., 0.04, 0., 0.],
            ['2SP192', 'Single Jersey', 32, 28, 175, 1., 0., 0., 0., 0.],
            ['2SP206', 'Single Jersey', 32, 28, 175, 1., 0., 0., 0., 0.],
            ['2SP247', 'Single Jersey', 32, 22, 200, 1., 0., 0., 0., 0.],
            ['2SP309', 'Single Jersey', 32, 28, 170, 1., 0., 0., 0., 0.],
            ['2SP388', 'Single Jersey', None, None, 170, 0., 0., 0., 1., 0.],
            ['2SP464', 'Single Jersey', 34, 14, 235, 0.45, 0., 0., 0., 0.55]
       ]
data
```

```python
#fabricDetailFrame is being created
fabricDetailFrame = pd.DataFrame(data = data, columns = fabricDetailColumn)
fabricDetailFrame
```

```python
#fabircDetailFrame is being finalized
tempDf = pd.DataFrame(dataFrame.groupby(['FabricCode', 'FabricDetail']).size()).reset_index().drop([0], axis = 1)
fabricDetailFrame = tempDf.merge(fabricDetailFrame, on='FabricCode', how='inner')
fabricDetailFrame
```

```python
#Information of fabricDetailFrame is being printed
fabricDetailFrame.info()
```

```python
#Statistical information of fabricDetailFrame is being printed
fabricDetailFrame.describe()
```

```python
#Null values are being filled with the median of the relevant columns
fabricDetailFrame = fabricDetailFrame.fillna(fabricDetailFrame.median(numeric_only=True))
fabricDetailFrame
```

```python
#Pus and Fine columns are being converted to int64 data type
fabricDetailFrame['Pus'] = fabricDetailFrame['Pus'].astype(np.int64)
fabricDetailFrame['Fine'] = fabricDetailFrame['Fine'].astype(np.int64)
fabricDetailFrame
```

```python
#Information of fabricDetailFrame is being printed
fabricDetailFrame.info()
```

```python
#Statistical information of fabricDetailFrame is being printed
fabricDetailFrame.describe()
```

```python
#joining dataFrame and fabricDetailFrame with inner join on FabricCode 
#will drop the rows with FabricCode values in the dataFrame that are not in the fabricDetailFrame
#In other words, this process will remove the product
#produced with the 2PK070 coded fabric, which was decided not to be used, from the dataFrame
dataFrame = dataFrame.merge(fabricDetailFrame[['FabricCode']], on='FabricCode', how='inner')
dataFrame
```

```python
#FabricDetail column is being removed from dataFrame
dataFrame.drop(['FabricDetail'], axis = 1, inplace = True)
dataFrame
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataframe is being printed
dataFrame.describe()
```

```python
#Since FabricCode is can be used as id for FabricDetail, there is no need to create a foreign key for FabricDetail

#dataFrame and relational fabricDetailFrame are being saved as pkl files
dataFrame.to_pickle("../Data/WithRelationalTables/FabricWasteFabricDetail.pkl")
fabricDetailFrame.to_pickle("../Data/WithRelationalTables/FabricDetail.pkl")
```
