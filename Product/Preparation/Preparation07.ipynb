# Seventh Data Preparation Process

In this notebook file, the dataset provided by the customer will be eddited for general purposes.

Output features will be examined and processed.

```python
#Libraries are being imported
import math
import pandas as pd
```

```python
#Library version is being printed
print('pandas Version: ' + pd.__version__)
```

```python
#A pandas dataframe named as dataFrame is being created by reading the data pkl file created last process
dataFrame = pd.read_pickle("../Data/WithRelationalTables/FabricWasteFabricDetail.pkl")
dataFrame.head()
```

```python
#Information of dataframe is being printed
dataFrame.info()
```

```python
#Statistical information of dataframe is being printed
dataFrame.describe()
```

```python
#SecondQuality Values recalculated based on Errors
dataFrame['CalculatedSecondQuality'] = dataFrame.apply(lambda row : row['PrintError'] + row['SewingError'] + 
                                                       row['FabricStain'] + row['FabricError'] +
                                                       row['EmbroideryError'] + row['MeasureError'], axis = 1)
dataFrame
```

```python
#Wrong SecondQuality Values are being checked
dataFrame[dataFrame['SecondQuality'] != dataFrame['CalculatedSecondQuality']].shape
```

```python
#Since there is no wrong value on SecondQuality, CalculatedSecondQuality column is being removed from dataFrame
dataFrame.drop('CalculatedSecondQuality', axis = 1, inplace = True)
```

```python
#Defect quantity is being calculated by summation of Repair and SecondQuality features
dataFrame.insert(loc=11, column='Defect', value = dataFrame['Repair'] + dataFrame['SecondQuality'])
dataFrame.head()
```

```python
#First row of dataFrame is being printed
dataFrame.iloc[0]
```

```python
#Within the scope of this work it is desired to predict 
#at least how many products should be produced for the orders together with the wastage

#If the situation is examined through the example in the first row:
#405 products were produced, 26 of these products were defective products,
#20 of them were repaired successfully but 6 of them were reserved as SecondQuality
#Ordered: 375, Delivered: 399
#The aim of this work is to bring the number of delivered products as close as possible to the number of ordered products

#For this process the some calculations should be done
#PrintErrorRate is being calculated for each row
dataFrame['PrintErrorRate'] = dataFrame.apply(lambda row : row['PrintError'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#SewingErrorRate is being calculated for each row
dataFrame['SewingErrorRate'] = dataFrame.apply(lambda row : row['SewingError'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#FabricStainRate is being calculated for each row
dataFrame['FabricStainRate'] = dataFrame.apply(lambda row : row['FabricStain'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#FabricErrorRate is being calculated for each row
dataFrame['FabricErrorRate'] = dataFrame.apply(lambda row : row['FabricError'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#EmbroideryErrorRate is being calculated for each row
dataFrame['EmbroideryErrorRate'] = dataFrame.apply(lambda row : row['EmbroideryError'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#MeasureErrorRate is being calculated for each row
dataFrame['MeasureErrorRate'] = dataFrame.apply(lambda row : row['MeasureError'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#SecondQualityRate is being calculated for each row
dataFrame['SecondQualityRate'] = dataFrame.apply(lambda row : row['SecondQuality'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#The DefectRate can be used for less risky predictions, while other rates can be used for more risky predictions
#DefectRate is being calculated for each row
dataFrame['DefectRate'] = dataFrame.apply(lambda row : row['Defect'] / row['Quantity'], axis = 1)
dataFrame.head()
```

```python
#The Delivery column is an integer type attribute that contains
#the number of products that has quality which can be delivered to the customer
#The Delivery feature can be calculated by subtracting the SecondQuality from Quantity
#and since prediction can be done on SecondQuality Feature
#Delivery is being removed from the dataset
dataFrame.drop('Delivery', axis = 1, inplace = True)
dataFrame.head()
```

```python
#Since required rates have been calculated, Quantity, Defect, Repair and SecondQuality columns can be removed now
dataFrame.drop('Quantity', axis = 1, inplace = True)
dataFrame.drop('Defect', axis = 1, inplace = True)
dataFrame.drop('Repair', axis = 1, inplace = True)
dataFrame.drop('PrintError', axis = 1, inplace = True)
dataFrame.drop('SewingError', axis = 1, inplace = True)
dataFrame.drop('FabricStain', axis = 1, inplace = True)
dataFrame.drop('FabricError', axis = 1, inplace = True)
dataFrame.drop('EmbroideryError', axis = 1, inplace = True)
dataFrame.drop('MeasureError', axis = 1, inplace = True)
dataFrame.drop('SecondQuality', axis = 1, inplace = True)
dataFrame.head()
```

```python
#New SecondQuality feature should be calculated according to the OrderQuantity

#Equation Example for these calculations:

#OrderQuantity = Quantity - Quantity * SecondQualityRate
#Quantity * (1 - SecondQualityRate) = OrderQuantity
#Quantity = OrderQuantity / (1 - SecondQualityRate)
#SecondQuality = Quantity - OrderQuantity
#SecondQuality = (OrderQuantity / (1 - SecondQualityRate)) - OrderQuantity

#SecondQuality = Ceil((OrderQuantity / (1 - SecondQualityRate)) - OrderQuantity)
#The ceil function is used to round the result to upper integer

#New SecondQuality features are being calculated according to the number of orders
dataFrame['SecondQuality'] = dataFrame.apply(lambda row : 
                            math.ceil((row['OrderQuantity'] / (1 - row['SecondQualityRate'])) - row['OrderQuantity']), axis = 1)
dataFrame.head()
```

```python
#New error features should be calculated according to the OrderQuantity and SecondQuality

#Equation Example for these calculations:

#PrintErrorRate = PrintError / Quantity
#PrintError = PrintErrorRate * Quantity
#Quantity = OrderQuantity + SecondQuality
#PrintError = PrintErrorRate * (OrderQuantity + SecondQuality)

#PrintError = Ceil(PrintErrorRate * (OrderQuantity + SecondQuality))
#The ceil function is used to round the result to upper integer

#New PrintError features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['PrintError'] = dataFrame.apply(lambda row : 
                            math.ceil(row['PrintErrorRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#New SewingError features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['SewingError'] = dataFrame.apply(lambda row : 
                           math.ceil(row['SewingErrorRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#New FabricStain features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['FabricStain'] = dataFrame.apply(lambda row : 
                            math.ceil(row['FabricStainRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#New FabricError features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['FabricError'] = dataFrame.apply(lambda row : 
                            math.ceil(row['FabricErrorRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#New EmbroideryError features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['EmbroideryError'] = dataFrame.apply(lambda row : 
                        math.ceil(row['EmbroideryErrorRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#New MeasureError features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['MeasureError'] = dataFrame.apply(lambda row : 
                            math.ceil(row['MeasureErrorRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```

```python
#Because a rounding function is used in the calculations,
#the SecondQuality value may not equal the sum of the error values in some rows

#SecondQuality Values recalculated based on Errors
dataFrame['CalculatedSecondQuality'] = dataFrame.apply(lambda row : row['PrintError'] + row['SewingError'] + 
                                                       row['FabricStain'] + row['FabricError'] +
                                                       row['EmbroideryError'] + row['MeasureError'], axis = 1)
dataFrame
```

```python
#Number of rows with SecondQuality not equal to CalculatedSecondQuality are being checked
dataFrame[dataFrame['SecondQuality'] != dataFrame['CalculatedSecondQuality']].shape
```

```python
#Number of rows with SecondQuality less than CalculatedSecondQuality are being checked

#As expected, all CalculatedSecondQuality values are greater than SecondQuality values
#since the CalculatedSecondQuality feature specifies that more products should be produced,
#this values can be used for more assured estimations

dataFrame[dataFrame['SecondQuality'] < dataFrame['CalculatedSecondQuality']].shape
```

```python
#New Defect features are being calculated according to the OrderQuantity and SecondQuality
dataFrame['Defect'] = dataFrame.apply(lambda row : 
                               math.ceil(row['DefectRate'] * (row['OrderQuantity'] + row['SecondQuality'])), axis = 1)
dataFrame.head()
```
