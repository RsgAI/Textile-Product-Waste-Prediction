# Polynomial Regression

In this notebook file, a Polynomial Regression model will be trained with FabricWaste data for predict SecondQualityRate

```python
#Libraries are being imported
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import __version__ as matplotlibVersion

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import __version__ as sklearnVersion

from joblib import dump, __version__ as joblibVersion
```

```python
#Library versions are being printed
print('numpy Version: ' + np.__version__)
print('pandas Version: ' + pd.__version__)
print('matplotlib Version: ' + matplotlibVersion)
print('sklearn Version: ' + sklearnVersion)
print('joblib Version: ' + joblibVersion)
```

```python
#Constant is being defined
MODEL_NAME = 'Polynomial Regression'
```

```python
#trainingFrame is being created by reading a pkl file
trainingFrame = pd.read_pickle("../../../Data/DataSplit/Training.pkl")
trainingFrame
```

```python
#Information of trainingFrame is being printed
trainingFrame.info()
```

```python
#Statistical information of trainingFrame part to use is being printed
trainingFrame.loc[:, : 'DefectRate'].describe()
```

```python
#trainingX is being created with input features of trainingFrame
trainingX = trainingFrame.loc[:, : 'PC22'].values
trainingX.shape
```

```python
#trainingY is being created with SecondQualityRate feature of trainingFrame
#Rates in the 0-1 range are being multiplied by 100 to get them into the 0-100 range
trainingY = trainingFrame['SecondQualityRate'].values * 100
trainingY.shape
```

```python
#Order 2 PolynomialFeatures model is being trained with trainingX data and updated with trainingX Polynomial Feature values
polyFeature = PolynomialFeatures(degree=2)
trainingX = polyFeature.fit_transform(trainingX)
trainingX.shape
```

```python
#A Ridge regression model is being trained with Polynomial Feature
regressor = Ridge().fit(trainingX, trainingY)
```

```python
#predictions is being made by trained regressor on trainingX data
predictions = regressor.predict(trainingX)
predictions.shape
```

```python
#resultFrame based on predictions and real values is being created
resultDict = {'Model' : MODEL_NAME,
              'Data' : 'Training',
              'MeanAbsoluteError' : [mean_absolute_error(predictions, trainingY)], 
              'MeanSquaredError' : [mean_squared_error(predictions, trainingY)]}
resultDict['SqrtMeanSquaredError'] = np.sqrt(resultDict['MeanSquaredError'])
resultFrame = pd.DataFrame(resultDict)
resultFrame
```

```python
#validationFrame is being created by reading a pkl file
validationFrame = pd.read_pickle("../../../Data/DataSplit/Validation.pkl")
validationFrame
```

```python
#Information of validationFrame is being printed
validationFrame.info()
```

```python
#Statistical information of validationFrame part to use is being printed
validationFrame.loc[:, : 'DefectRate'].describe()
```

```python
#validationX is being created with input features of validationFrame and transformed into Polynomial Feature form
validationX = validationFrame.loc[:, : 'PC22'].values
validationX = polyFeature.transform(validationX)
validationX.shape
```

```python
#validationY is being created with SecondQualityRate feature of validationFrame
#Rates in the 0-1 range are being multiplied by 100 to get them into the 0-100 range
validationY = validationFrame['SecondQualityRate'].values * 100
validationY.shape
```

```python
#predictions are being made by trained regressor on validationX data
predictions = regressor.predict(validationX)
predictions.shape
```

```python
#resultFrame based on predictions and real values is being created
resultDict = {'Model' : MODEL_NAME,
              'Data' : 'Validation',
              'MeanAbsoluteError' : [mean_absolute_error(predictions, validationY)], 
              'MeanSquaredError' : [mean_squared_error(predictions, validationY)]}
resultDict['SqrtMeanSquaredError'] = np.sqrt(resultDict['MeanSquaredError'])
resultFrame = pd.concat([resultFrame, pd.DataFrame(resultDict)]).reset_index(drop = True)
resultFrame
```

```python
#Chart to show the closeness of the SecondQualityRate predictions to the true values is being plaotted
plt.figure(figsize = (12, 6))
plt.title('SecondQualityRate')
plt.scatter(validationY, predictions)
plt.plot(validationY, validationY, 'r')
plt.legend(['Predictions', 'LineOfRealValues'])
```

```python
predictions[predictions < 0] = 0
predictions[predictions > 8] = 8
predictions
```

```python
#resultFrame based on predictions and real values is being created
resultDict = {'Model' : MODEL_NAME,
              'Data' : 'Adjusted Validation',
              'MeanAbsoluteError' : [mean_absolute_error(predictions, validationY)], 
              'MeanSquaredError' : [mean_squared_error(predictions, validationY)]}
resultDict['SqrtMeanSquaredError'] = np.sqrt(resultDict['MeanSquaredError'])
resultFrame = pd.concat([resultFrame, pd.DataFrame(resultDict)]).reset_index(drop = True)
resultFrame
```

```python
#Chart to show the closeness of the SecondQualityRate predictions to the true values is being plaotted
plt.figure(figsize = (12, 6))
plt.title('SecondQualityRate')
plt.scatter(validationY, predictions)
plt.plot(validationY, validationY, 'r')
plt.legend(['Predictions', 'LineOfRealValues'])
```
